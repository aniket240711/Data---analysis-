```python
def standardize_job_title(title):
    title = str(title).lower()
    if 'data scientist' in title or 'data science' in title:
        if 'machine learning' in title or 'ml' in title:
            return 'Machine Learning Scientist'
        elif 'manager' in title:
            return 'Data Science Manager'
        elif 'sr' in title or 'senior' in title:
            return 'Senior Data Scientist'
        elif 'jr' in title or 'junior' in title or 'associate' in title:
            return 'Junior Data Scientist'
        return 'Data Scientist'
    elif 'data engineer' in title:
        if 'sr' in title or 'senior' in title:
            return 'Senior Data Engineer'
        return 'Data Engineer'
    elif 'data analyst' in title:
        if 'sr' in title or 'senior' in title:
            return 'Senior Data Analyst'
        return 'Data Analyst'
    elif 'machine learning engineer' in title or 'ml engineer' in title or 'machine learning scientist' in title:
        return 'Machine Learning Engineer'
    elif 'research scientist' in title or 'computational scientist' in title:
        return 'Research Scientist'
    elif 'business intelligence' in title or 'bi analyst' in title:
        return 'Business Intelligence Analyst'
    elif 'software engineer' in title:
        return 'Software Engineer'
    elif 'analytics manager' in title:
        return 'Analytics Manager'
    elif 'director' in title:
        return 'Director of Data Science'
    elif 'manager' in title:
        return 'Manager (Other)' # For other manager roles not specific to DS
    elif 'scientist' in title:
        return 'Scientist (Other)' # For other scientist roles
    else:
        return 'Other'

df1['Job Title'] = df1['Job Title'].apply(standardize_job_title)
print("Job Title column standardized.")
```

```python
state_abbreviations = {
    'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA',
    'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'Florida': 'FL', 'Georgia': 'GA',
    'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA',
    'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
    'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS', 'Missouri': 'MO',
    'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH', 'New Jersey': 'NJ',
    'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH',
    'Oklahoma': 'OK', 'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC',
    'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT',
    'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY',
    'District of Columbia': 'DC'
}

def extract_state(location):
    if pd.isna(location) or location in ['-1', 'United States', 'Remote'] or 'Switzerland' in location or 'UK' in location:
        return np.nan
    location = str(location).strip()

    # Handle 'City, State Abbreviation' format
    if ',' in location:
        state_part = location.split(',')[-1].strip()
        if state_part in state_abbreviations.values():
            return state_part
        elif state_part in state_abbreviations.keys():
            return state_abbreviations[state_part]

    # Handle full state names or single abbreviations
    if location in state_abbreviations.keys():
        return state_abbreviations[location]
    if location in state_abbreviations.values():
        return location

    return np.nan

df1['Job State'] = df1['Location'].apply(extract_state)
df1['Headquarters State'] = df1['Headquarters'].apply(extract_state)

print("Extracted 'Job State' and 'Headquarters State' columns.")
```

```python
def parse_employee_size(size_str):
    if pd.isna(size_str) or size_str == 'Unknown':
        return np.nan, np.nan
    elif '+' in size_str:
        min_val = float(size_str.replace('+', ''))
        return min_val, np.inf # Using numpy.inf for '10000+'
    elif '-' in size_str:
        min_val, max_val = map(float, size_str.split('-'))
        return min_val, max_val
    else:
        return np.nan, np.nan # Handle any other unexpected format

# Apply the function to create new Min_Employees and Max_Employees columns
df1[['Min_Employees', 'Max_Employees']] = df1['Size(employees)'].apply(lambda x: pd.Series(parse_employee_size(x)))

print("Min_Employees and Max_Employees columns created from 'Size(employees)'.")
```
