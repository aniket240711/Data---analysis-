Question 1 : Explain the fundamental differences between DDL, DML, and DQL
commands in SQL. Provide one example for each type of command.


DDL (Data Definition Language): These commands are used to define, modify, or delete the structure of a database, including the creation and modification of tables, indexes, and users. DDL commands deal with the database schema.

Example (DDL): CREATE TABLE Customers (CustomerID INT PRIMARY KEY, CustomerName VARCHAR(255), Email VARCHAR(255));
DML (Data Manipulation Language): These commands are used for managing data within schema objects. They allow you to add, retrieve, modify, or delete data in a database.

Example (DML): INSERT INTO Customers (CustomerID, CustomerName, Email) VALUES (1, 'Alice Smith', 'alice.smith@example.com');
DQL (Data Query Language): This language is used specifically for querying the database and retrieving data based on certain criteria. While often grouped under DML, DQL is distinct because it does not modify the database.

Example (DQL): SELECT CustomerName, Email FROM Customers WHERE CustomerID = 1;
In essence:

DDL handles the structure.
DML handles the data (insertion, updating, deletion).
DQL handles retrieving data.


Question 2 : What is the purpose of SQL constraints? Name and describe three common types
of constraints, providing a simple scenario where each would be useful.


SQL constraints are rules enforced on data columns in a table to limit the type of data that can be inserted or updated into a table. The primary purpose of constraints is to maintain the data integrity and accuracy within a database. By applying constraints, you ensure that the data adheres to predefined business rules and standards.

Here are three common types of SQL constraints:

PRIMARY KEY:

Description: A PRIMARY KEY constraint uniquely identifies each record in a table. It must contain unique values, and it cannot contain NULL values. A table can only have one primary key, which can consist of single or multiple columns.
Scenario: In a Customers table, CustomerID would be an ideal primary key. Each customer should have a unique ID, and this ID is crucial for identifying and referencing individual customer records across the database.
FOREIGN KEY:

Description: A FOREIGN KEY constraint is a field (or collection of fields) in one table that refers to the PRIMARY KEY in another table. It establishes a link between two tables, ensuring referential integrity by preventing actions that would destroy links between related tables.
Scenario: If you have an Orders table that stores customer orders, CustomerID in the Orders table could be a foreign key referencing the CustomerID (primary key) in the Customers table. This ensures that every order placed is associated with an existing customer.
NOT NULL:

Description: The NOT NULL constraint ensures that a column cannot have a NULL value. It enforces that a column must always contain a value, preventing blank entries in critical fields.
Scenario: In a Products table, the ProductName column should often be set as NOT NULL. It wouldn't make sense to have a product record without a name, so this constraint ensures that every product has a defined name.


Question 3 : Explain the difference between LIMIT and OFFSET clauses in SQL. How
would you use them together to retrieve the third page of results, assuming each page
has 10 records?

The LIMIT and OFFSET clauses are fundamental for controlling the number of rows returned by a SQL query and for implementing pagination.

LIMIT Clause:

Purpose: The LIMIT clause is used to restrict the number of rows returned by the query to a specified maximum value. It's often used when you only need a subset of the results, such as the top N records.
Example: SELECT * FROM Products LIMIT 5; (This would return the first 5 rows from the Products table).
OFFSET Clause:

Purpose: The OFFSET clause is used to skip a specified number of rows from the beginning of the result set before applying the LIMIT. It's particularly useful for pagination, allowing you to fetch subsequent 'pages' of data.
Example: SELECT * FROM Products OFFSET 10; (This would skip the first 10 rows and return all subsequent rows from the Products table).
Using LIMIT and OFFSET for Pagination:

To retrieve a specific page of results, you combine LIMIT to set the number of records per page and OFFSET to skip the records from previous pages.

Scenario: Retrieving the third page of results, with each page having 10 records.

Here's how you calculate the OFFSET:

Page 1: LIMIT 10 OFFSET 0 (skips 0 records, takes the first 10)
Page 2: LIMIT 10 OFFSET 10 (skips the first 10 records, takes the next 10)
Page 3: LIMIT 10 OFFSET 20 (skips the first 20 records, takes the next 10)
In general, for page N and M records per page, the OFFSET would be (N - 1) * M.





Question 4 : What is a Common Table Expression (CTE) in SQL, and what are its main
benefits? Provide a simple SQL example demonstrating its usage.

 Common Table Expression (CTE) is a temporary, named result set that you can reference within a single SQL statement (SELECT, INSERT, UPDATE, or DELETE). It's essentially a named subquery that makes complex queries more readable and manageable.

CTEs are defined using the WITH clause, followed by the CTE name, and then its definition.

Main Benefits of CTEs:

Readability and Maintainability: CTEs break down complex queries into logical, self-contained, named blocks. This makes the SQL code easier to read, understand, and maintain, especially for queries involving multiple joins, subqueries, or intricate logic.
Recursion (Recursive CTEs): CTEs are crucial for handling recursive queries, such as traversing hierarchical data (e.g., organizational charts, bill of materials). This allows a CTE to refer to itself.
Improved Organization: By defining logical steps in a query, CTEs help organize your SQL, making it more structured and modular.
Reusability within a Single Query: Although a CTE is temporary and exists only for the duration of its query, you can reference the same CTE multiple times within that single query. This can simplify code and ensure consistent logic across different parts of the query.
Simulating Views: CTEs can act like temporary views for the duration of a single query, which is useful when you need a custom dataset for a specific operation without creating a permanent view object.

Question 5 : Describe the concept of SQL Normalization and its primary goals. Briefly
explain the first three normal forms (1NF, 2NF, 3NF).


SQL Normalization is a systematic process of organizing the columns and tables of a relational database to minimize data redundancy and improve data integrity. It involves decomposing tables into smaller, related tables and defining relationships between them.

Primary Goals of SQL Normalization:

Eliminate Data Redundancy: Avoid storing the same piece of data in multiple places, which can lead to inconsistencies and wasted storage space.
Improve Data Integrity: Ensure that data is accurate and consistent throughout the database by preventing update, insertion, and deletion anomalies.
Enhance Database Efficiency: Make queries more efficient by reducing the amount of data the database system needs to process.
Simplify Database Design: Create a more logical and understandable database structure.
Brief Explanation of the First Three Normal Forms (1NF, 2NF, 3NF):

First Normal Form (1NF):

Rule: A table is in 1NF if it contains no repeating groups or multi-valued attributes, and each column contains atomic (indivisible) values. This means each cell in the table should hold only one value, and there should be no repeating columns for a single record.
