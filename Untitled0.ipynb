## Extract Winning Margin from 'result' Column



```python
import re

def extract_winning_margin(result_string):
    if not isinstance(result_string, str):
        return None, None

    runs_pattern = r'won by (\d+) runs'
    wickets_pattern = r'won by (\d+) wkts|won by (\d+) wickets'

    runs_match = re.search(runs_pattern, result_string)
    wickets_match = re.search(wickets_pattern, result_string)

    if runs_match:
        return int(runs_match.group(1)), None
    elif wickets_match:
        # Check both patterns for wickets (wkts or wickets)
        if wickets_match.group(1):
            return None, int(wickets_match.group(1))
        elif wickets_match.group(2):
            return None, int(wickets_match.group(2))
    return None, None

df[['winning_runs', 'winning_wickets']] = df['result'].apply(lambda x: pd.Series(extract_winning_margin(x)))
df['winning_runs'].fillna(0, inplace=True)
df['winning_wickets'].fillna(0, inplace=True)
df['winning_runs'] = df['winning_runs'].astype(int)
df['winning_wickets'] = df['winning_wickets'].astype(int)

print("First few rows of the DataFrame with new 'winning_runs' and 'winning_wickets' columns:")
display(df.head())

print("\nData types of 'winning_runs' and 'winning_wickets' columns:")
print(df[['winning_runs', 'winning_wickets']].dtypes)
```
