### Matplotlib Pie Chart Demo (GitHub Visible)

```python
import matplotlib.pyplot as plt

# The variables `labels`, `sizes`, `colors`, and `explode` are already
# defined in the current kernel state from a previous execution.
#For reference, they are:
labels = ['Apple', 'Banana', 'Orange', 'Grapes']
sizes = [30, 25, 20, 25]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # Explode the 1st slice (Apple)

plt.figure(figsize=(7, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Fruits (Demo)')
plt.tight_layout()
plt.show()
```
