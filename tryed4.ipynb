```python
import pandas as pd
import plotly.express as px

# Sample data for the pie chart
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Value': [30, 20, 15, 35]
}
df_pie = pd.DataFrame(data)

# Create the pie chart
fig = px.pie(df_pie, values='Value', names='Category', title='Sample Pie Chart')

# Save the figure as a static image (e.g., PNG) for GitHub visibility
# Make sure kaleido is installed for image export: !pip install -U kaleido
fig.write_image("sample_pie_chart.png")

# Display the chart in the notebook
fig.show()
```

The code above first creates a sample Pandas DataFrame. Then, it uses `plotly.express.pie()` to generate a pie chart, where `values` specifies the size of each slice and `names` provides the labels for the slices. Finally, `fig.write_image()` saves the chart as a PNG file, which can be easily displayed on GitHub. The `fig.show()` command displays the interactive plot in the Colab notebook.
